## Punto 1

```python
def f(x,y):
    return x + 2*y

def Solucion(x):
    return -0.5*x - (1-np.exp(2*x))/4.
```

```python
def RungeKutta_Third(x):
    h = x[1] - x[0]
    y = np.zeros_like(x)
    y[0] = 0
    for i in range(1, len(x)):
        K1 = f(x[i-1],y[i-1])
        K2 = f(x[i-1] + 0.5*h, y[i-1] + 0.5*K1*h)
        K3 = f(x[i-1] + 0.5*h, y[i-1] - K1*h + 2*K2*h)
        
        y[i] = y[i - 1] + (h/6)*(K1 + 4*K2 + K3)
        
    return y

def RungeKutta_Fourth(x):
    h = x[1] - x[0]
    y = np.zeros_like(x)
    y[0] = 0
    for i in range(1,len(x)):
        K1 = f(x[i-1],y[i-1])
        K2 = f(x[i-1] + 0.5*h, y[i-1] + 0.5*K1*h)
        K3 = f(x[i-1] + 0.5*h, y[i-1] + 0.5*K2*h)
        K4 = f(x[i-1] + h, y[i - 1] + K3*h)
        
        y[i] = y[i-1] + (1/6)*(K1 + 2*K2 + 2*K3 + K4)*h
    
    return y
```

```python
x = np.linspace(0, 1, 10)
y_3th = RungeKutta_Third(x)
y_4th = RungeKutta_Fourth(x)
y = Solucion(x)
```

```python
def fc(z):
    return 1j*z

def Complex_Third():
    global theta
    global h_
    global f
    
    theta += h_
    K1 = fc(f)
    K2 = fc(f + 0.5*K1*h_)
    K3 = fc(f - K1*h_ + 2*K2*h_)
    
    f += (h_/6)*(K1 + 4*K2 + K3)
    
    
def Complex_Fourth():
    global theta
    global h_
    global f
    
    theta += h_
    K1 = fc(f)
    K2 = fc(f + 0.5*K1*h_)
    K3 = fc(f + 0.5*K2*h_)
    K4 = fc(f + K3*h_)
        
    f += (h_/6)*(K1 + 2*K2 + 2*K3 + K4)
```

```python
theta = 0.
N = 15
Vueltas = 1
h_ = Vueltas * 2*np.pi/N
f = 1

SolucionRK3 = np.array([theta,f])

for i in range(N):
    Complex_Third()
    SolucionRK3 = np.vstack([SolucionRK3, np.array([theta,f])])

ReRK3 = np.real(SolucionRK3[:,1])
ImRK3 = np.imag(SolucionRK3[:,1])

SolucionRK4 = np.array([theta,f])

for i in range(N):
    Complex_Fourth()
    SolucionRK4 = np.vstack([SolucionRK4, np.array([theta,f])])

ReRK4 = np.real(SolucionRK4[:,1])
ImRK4 = np.imag(SolucionRK4[:,1])
```

## Punto 2

### Parte a

El método de Verlet se obtiene a partir de la suma por izquierda y derecha de la aproximación de la función por expación de Taylor que son:

$x(t + \delta t) = x(t) + v(t)\delta t + \frac{(\delta t)^{2}}{2}a(t) + \mathcal{O}[(\delta t)^{3}]$ 

$x(t - \delta t) = x(t) - v(t)\delta t + \frac{(\delta t)^{2}}{2}a(t) - \mathcal{O}[(\delta t)^{3}]$

Suma cuyo resultado es:

$x(t + \delta t) + x(t - \delta t) = 2x(t) + a(t)(\delta t)^{2}$

Que al discretizar es igual a

$x_{n + 1} + \epsilon_{n + 1} + x_{n - 1} + \epsilon_{n - 1} = 2x_{n} + 2\epsilon_{n} + a(x_{n} + \epsilon_{n})h^{2}$

Que si aproximamos el error de la aceleración con una expansión de Taylor tenemos que:

$x_{n + 1} + \epsilon_{n + 1} + x_{n - 1} + \epsilon_{n - 1} = 2x_{n} + 2\epsilon_{n} + ax_{n}h^{2} + a'_{n} \epsilon_{n}h^{2}$

Finalmente,

$\epsilon_{n + 1} + \epsilon_{n - 1} = 2\epsilon_{n} + a'_{n}\epsilon_{n}h^{2}$

Que al reorganizar términos nos queda,

$\epsilon_{n + 1} = \left(2 + a'_{n} h^{2}\right)\cdot \epsilon_{n} - \epsilon_{n - 1}$

### Parte b

Para un oscilador armónico se tiene que $a = -\omega^{2}\cdot x$, luego $a'_{x}=-\omega^{2}$. Por lo que

$\epsilon_{n + 1} = \left(2 + a'_{x}h^{2}\right)\cdot \epsilon_{n} - \epsilon_{n - 1}$

$\epsilon_{n + 1} = \left(2 - \omega^{2}h^{2}\right)\cdot \epsilon_{n} - \epsilon_{n - 1}$

Sabiendo que $2R = h^{2}\omega^{2}$

$\epsilon_{n + 1} = 2 \cdot (1 - R) \cdot \epsilon_{n} + \epsilon_{n - 1}$


### Parte c



Si asumimos la suposición de la función potencia, tenemos que

$\epsilon_{0}\lambda^{n + 1} = 2 \cdot (1 - R) \cdot \epsilon_{0}\lambda^{n} + \epsilon_{0}\lambda^{n - 1}$

Si dividimos por el factor más pequeño $\epsilon_{0}\lambda^{n - 1}$, tenemos que:

$\lambda^{2} = 2 \cdot (1 - R) \cdot \lambda + 1$

Que es una ecuación ecuación cuadrática de $\lambda$

$\lambda^{2} - 2 \cdot (1 - R) \cdot \lambda - 1 = 0$

Al resolverla tenemos que:

$\lambda_{\pm} = \frac{2(1 - R) \pm \sqrt{4(1 - R)^{2} - 4}}{2}$

Luego,

$\lambda_{\pm} = 1 - R \pm \sqrt{R^{2} - 2R}$


### Parte d

Tenemos que:

$\left|\lambda_{\pm}\right| = \left|1 - R \pm \sqrt{R^{2} - 2R}\right| \leq 1 \quad$ 

Luego 

$-1 \leq 1 - R \pm \sqrt{R^{2} - 2R} < 1$

$ - 2 \leq  - R \pm \sqrt{R^{2} - 2R} \leq 0$

$ R - 2 \leq  \pm \sqrt{R^{2} - 2R} \leq R $ 

$R^{2} \leq  \left(\pm \sqrt{R^{2} - 2R}\right)^{2} \leq (R - 2)^{2}$

$R^{2} \leq  R^{2} - 2R \leq R^{2} - 4R + 4$

Si nos quedamos con el lado derecho de la desigualdad

$R^{2} - 2R \leq R^{2} - 4R + 4$

$2R \leq 4 \quad $

Sabemos que $2R = h^{2}\omega^{2}$, entonces

$h \leq \frac{2}{\omega}$

### Parte e

```python
x0 = np.array([1., 1.])

def Harmonic_Oscillator(x):
    return np.cos(np.pi*x) + (1/np.pi)*np.sin(np.pi*x)

def Verlet_Integration(x0, time, step):
    t = np.arange(0, time, step)
    x_computed = np.zeros_like(t)
    x_real = np.zeros_like(t)
    x_computed[0] = x0[0]
    
    for i in tqdm(range(0, len(t) - 1), desc='Running simulation', unit=' Steps' ):
        a = -(np.pi**2) * x_computed[i]
        if i == 0:
            v_ = x0[1] + step*a
            x_computed[i + 1] = x_computed[i] + step*v_
        else:
            x_computed[i + 1] = 2*x_computed[i] - x_computed[i - 1] + a*(step**2)
        
        x_real[i] = Harmonic_Oscillator(t[i])
    return x_computed, x_real, t
```

```python
x_estable, x_real, t_estable = Verlet_Integration(x0, 15, 0.1)
```

## Punto 3

Tenemos que, que es una ecuación diferencial de Riccati:

$x^3y' = x^4y^2 - 2x^2y-1$

Luego

$ y' = xy^2 - \frac{2y}{x} - \frac{1}{x^3}$

Si tomamos que $y = y_1 + z$, tenemos entonces que

$ y'_1 + z' = x(y_1 + z)^2 - \frac{2}{x}(y_1 + z) - \frac{1}{x^3} $

$ y'_1 + z' = xy_1^2 + 2xy_1z + xz^2 - \frac{2y_1}{x} - \frac{2z}{x} - \frac{1}{x^3}.$

Si $y_{1}$ es solución particular de la ecuación con $y_{1}=x^-2$

$z' = xz^2$

Si hacemos un cambio de variable del tipo $z = u^{-1}$, luego $z' = u^{-2}u'$. Entonces,

$-u^{-2}u' = xu^{-2}$

$u = -\frac{x^2}{2} + c_1$ luego, $z = \left(-\frac{x^2}{2} + c_1\right)^{-1}$ 

Por lo que la solución general es 

$y = x^{-2} + \left(-\frac{x^2}{2} + c_1\right)^{-1}$

Que al tomar la condición inicial en la cual $y(\sqrt2)=0$ nos da que:

$y = x^{-2} - \left(\frac{x^2}{2} +1\right)^{-1}$

```python
FuncionR = lambda x: 1/x**2 - 1/(x**2/2 + 1)

def f(x,y):
    return x*y**2-2*y/x-1/x**3

def GetEuler2(f,x):
    dx = x[1] - x[0]
    y = np.zeros_like(x)
    y[0] = 0
    for i in range(1,len(x)):
        k1 = f(x[i-1],y[i-1])
        k2 = f(x[i],y[i-1]+k1*dx)
        
        y[i] = y[i-1]+.5*(k1+k2)*dx
    
    return y

x = np.linspace(np.sqrt(2),5,100)

y_sol = GetEuler2(f,x)

fig3 = plt.figure(figsize=(5,5))
ax3 = fig3.add_subplot(111)

ax3.set_title("Solución")

ax3.grid()
ax3.set_xlim(1,5)
ax3.set_ylim(-0.1,0.1)

ax3.vlines(np.sqrt(2),-0.1,0.1,linestyles="dashed",label="Condición inicial")
ax3.plot(x,y_sol,label="Solución numérica")
ax3.plot(x,FuncionR(x),label="Solución analítica")
ax3.legend()
```

## Punto 4

Para $q=1$, la solución es trivial

$u(t) = e^{t}$

Si $q<1$ tenemos una ecuaciónde Bernoulli, entonces:

Si $z = u^{1-q}$ entones $u = z^{\frac{1}{1-q}}$ y

$\frac{\text{d}u}{\text{d}t} = \frac{1}{1-q}z^{\frac{1}{1-q}-1}\frac{\text{d}z}{\text{d}t}.$

Luego

$ zu^{-1} = u^{-q} $

$\frac{z}{z^{\frac{1}{1-q}}} = (1-q)z^{1-\frac{1}{1-q}}\left( \frac{\text{d}z}{\text{d}t}\right)^{-1} $

$z = (1-q)z\left( \frac{\text{d}z}{\text{d}t}\right)^{-1} \\
\frac{\text{d}z}{\text{d}t} = 1-q $

$ z = (1-q)t+c_1 $

$u = ((1-q)t + c_1)^{\frac{1}{1-q}}$

Tomando que $u(0) = 1$ tenemos que  

$c_1 = 1.$

Finalmente

$u(t) = \left((1-q)t +1\right)^{\frac{1}{1-q}}$

```python
G, alpha, e, a = 4*np.pi**2, 1.1e-8, 0.205630,0.387098

def Aceleration(r):
    d = np.linalg.norm(r)
    return -(G/d**3)*(1 + alpha/(d**2))*r

def Verlet_Moded(x0, time, step):
    t = np.arange(0, time, step)
    
    x_computed = np.zeros((len(t), 2))
    v_computed = np.zeros((len(t), 2))
    a_computed =np.zeros((len(t), 2))
    
    x_computed[0] = x0[0]
    v_computed[0] = x0[1]
    
    angulo = np.array([])
    tiempo = np.array([])
    
    reduced = False
    
    for i in tqdm(range(len(t) - 1), desc='Running simulation', unit=' Steps' ):
        a_computed[i] = Aceleration(x_computed[i])
        if i == 0:
            v_computed[i + 1] = v_computed[i] + step*a_computed[i]
            x_computed[i + 1] = x_computed[i] + step*v_computed[i]
        else:
            x_computed[i + 1] = 2*x_computed[i] - x_computed[i - 1] + a_computed[i]*(step**2)
            a_computed[i + 1] = Aceleration(x_computed[i + 1])
            v_computed[i + 1] = v_computed[i] + (a_computed[i + 1] + a_computed[i])*(step/2)
            
            if np.linalg.norm(x_computed[i + 1]) < np.linalg.norm(x_computed[i]):
                reduced = True
            else:
                if reduced:
                    xp = x_computed[i + 1]
                    ang = np.arctan(xp[1]/ xp[0])*(180/np.pi)
                    angulo = np.append(angulo, ang)
                    tiempo = np.append(tiempo, t[i])#//(3600*24*365))
                    reduced = False 
                

    return angulo, tiempo
```

```python
x0 = np.array([np.array([a*(1 - e), 0]), np.array([0, np.sqrt(G*(1 + e)/(a*(1. - e)))])])
t = np.arange(0, 1, 0.1)
x = np.zeros((len(t), 2))
x[0] = x0[0]
```

```python
angulo, tiempo = Verlet_Moded(x0, 1, 1e-8)
```

```python
def GetRegression(x,y):
    mean_y = np.mean(y)
    mean_x = np.mean(x)
    p = np.sum((x-mean_x)*(y-mean_y))
    q = np.sum((x-mean_x)**2)
    m = p/q
    b = mean_y - m*mean_x
    return m,b
```

```python
fig = plt.figure(figsize = (5, 5))
v_precs, b = GetRegression(tiempo, angulo)
plt.scatter(tiempo, angulo)
plt.plot(tiempo, v_precs*tiempo + b, color = "red", label = r"Ω = %s ArcSecondsCentury"%(np.round(v_precs*360000, 2)))
plt.ylabel(r"$\Delta\varphi [Grados]$")
plt.xlabel("Tiempo [Años]")
plt.legend()
```
